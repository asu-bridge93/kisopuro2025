### 媒介中心性の変化を港ごとに可視化する

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

# 日本語フォントを設定（IPAexGothicを使用）

# 上位3つと下位3つを再作成
data_top3 = {
    "Index": ["Bangkok", "Qingdao", "Fangcheng Port"],
    "Country": ["Thailand", "China", "China"],
    "Betweenness Centrality Difference": [38.09425619712556, 25.462019650519096, 20.727368085514627]
}
top3 = pd.DataFrame(data_top3)

data_bottom3 = {
    "Index": ["Rongcheng", "Zhangzhou Port", "Shidao Port"],
    "Country": ["China", "China", "China"],
    "Betweenness Centrality Difference": [-0.6493835371099941, -0.6613882128843586, -15.515063440572703]
}
bottom3 = pd.DataFrame(data_bottom3)

highlight_df = pd.concat([top3, bottom3])

# 再度グラフを描画
plt.figure(figsize=(8, 6))
bars = plt.barh(highlight_df["Index"], highlight_df["Betweenness Centrality Difference"], color=["orange" if x > 0 else "blue" for x in highlight_df["Betweenness Centrality Difference"]])
plt.xlabel("媒介中心性の変化量")
plt.title("媒介中心性の変化（上位3港・下位3港）")
plt.axvline(0, color="black", linewidth=0.8)  # 0ライン
plt.gca().invert_yaxis()
plt.tight_layout()
# 文字サイズ変更
plt.xlabel("媒介中心性の変化量", fontsize=14)  # ラベルのフォントサイズ
plt.title("媒介中心性の変化（上位3港・下位3港）", fontsize=16)  # タイトルのフォントサイズ
plt.xticks(fontsize=14, color="black")  # x軸の目盛りのフォントサイズ
plt.yticks(fontsize=14, color="black")  # y軸の目盛りのフォントサイズ

plt.show()
```
